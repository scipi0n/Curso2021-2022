```python
!pip install rdflib 
from rdflib import Graph, Namespace, Literal
from rdflib.namespace import RDF, RDFS
g = Graph()
g.namespace_manager.bind('dbr', Namespace("http://transporteMad.example.org#"), override=False)
g.namespace_manager.bind('dbo', Namespace("http://transporteMad.example.org/Ontology/"), override=False)
g.parse("https://raw.githubusercontent.com/spardod/Pruebas2/main/output.nt", format="nt")
dbr =  Namespace("http://transporteMad.example.org#")
dbo = Namespace("http://transporteMad.example.org/Ontology/")
res = Namespace("http://transporteMad.example.org#")

from rdflib.plugins.sparql import prepareQuery
q1 = prepareQuery('''
  SELECT ?Numero WHERE {
    ?Numero rdf:type dbr:EstacionBiciMad.
    ?Numero dbr:barrio "SOL"
    }
    ''',
    initNs = {"rdfs": RDFS,"dbr": dbr,"rdf": RDF}
)

for r in g.query(q1):
  print(r)

q2 = prepareQuery('''
  SELECT ?Numero WHERE{
    ?Numero rdf:type dbr:ParadaTaxi.
    ?Numero dbr:NumeroPlazas "5"
  }

''',
initNs = {"rdfs": RDFS,"dbr": dbr,"rdf": RDF}
)

for r in g.query(q2):
  print(r)

q3 = prepareQuery('''
  SELECT ?Codigo WHERE{
    ?Codigo rdf:type dbr:EstacionCercanias.
    ?Codigo dbr:Poblacion "Alpedrete"
  }

''',
initNs = {"rdfs": RDFS,"dbr": dbr,"rdf": RDF}
)

for r in g.query(q3):
  print(r)
```
